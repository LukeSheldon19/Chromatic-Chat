```python
df2 = pd.read_csv('../data/text.csv')
df2 = df2.drop(columns=['Unnamed: 0'])
```

```python
df2.head()
```

```python
# Label mapping --> sadness (0), joy (1), love (2), anger (3), fear (4), and surprise (5)
```

```python
label_map = {
    0: "sadness",
    1: "joy",
    2: "love",
    3: "anger",
    4: "fear",
    5: "surprise"
}

df2["label"] = df2["label"].map(label_map)
```
